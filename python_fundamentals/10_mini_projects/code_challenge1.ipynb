Write a function which computes and prints the running average from the list
[7, 13, 17, 231, 12, 8, 3]

```python
def calculate_average(list_of_numbers): 
    count = 1
    total = 0
    for num in list_of_numbers:  
        total  += num   
        average = total / count
        count += 1
        yield average
    
num_list = [7,13,17,231,12,8,3]
for average in calculate_average(num_list):
    print(average)

'''next() function is used to retrieve the next item from the generator.
    In the above code, we have a generator function calculate_average that yields the average of a list of numbers.
    We can use the next() function to get the next average from the generator each time it is called.
    f1 = calculate_average(num_list) creates a generator object f1.
    print(next(f1)) retrieves the next average from the generator and prints it. Each time next() is called, it continues from where it last left off in the generator function, allowing us to get the next average in the sequence.
    print(next(f1,"No more averages")) will print the next average if available, or "No more averages" if the generator has been exhausted (i.e., all averages have been calculated and there are no more items to yield).
 
 '''
```
