# Dictionaries in Python
A dictionary is an ordered, mutable collection of key-value pairs.
- Key-value pairs — each item has a key and a value
- Keys must be unique
- Keys must be immutable (string, int, tuple)
- Values can be any datatype

```python
# Empty dictionary
my_dict = {}

# Dictionary with items
student = {
    "name": "Nivedita",
    "age": 40,
    "city": "Chennai",
    "is_student": True
}

print(student)
print(type(student))
```

## Accessing Values

```python
student = {
    "name": "Nivedita",
    "age": 40,
    "city": "Chennai"
}

# Access by key
print(student["name"])

# get() - safer way, returns None if key not found
print(student.get("age"))
print(student.get("phone"))        # returns None
print(student.get("phone", "N/A")) # returns default value
```
