# Sets in Python
A set is an unordered, mutable collection of unique items.
- Unordered — items have no fixed position
- Mutable — items can be added or removed
- No duplicate values allowed
- No indexing or slicing!

```python
# Empty set - must use set(), not {}
my_set = set()

# Set of integers
numbers = {1, 2, 3, 4, 5}

# Duplicates are automatically removed!
fruits = {"apple", "banana", "mango", "apple", "banana"}

print(numbers)
print(fruits)
print(type(fruits))
```

## Adding and Removing Items

```python
fruits = {"apple", "banana", "mango"}

# add
fruits.add("grape")
print(fruits)

# remove - throws error if item not found
fruits.remove("banana")
print(fruits)

# discard - no error if item not found
fruits.discard("orange")  # orange doesn't exist but no error!
print(fruits)
```

## Set Operations
Sets support mathematical operations like union, intersection and difference.

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# union - all items from both sets
print(set1.union(set2))

# intersection - only common items
print(set1.intersection(set2))

# difference - items in set1 but not in set2
print(set1.difference(set2))
```

## Membership Test
Sets are much faster than lists for checking if an item exists!

```python
fruits = {"apple", "banana", "mango", "grape"}

print("apple" in fruits)
print("orange" in fruits)
print(len(fruits))
```
