## GroupBy and Aggregations
Learning to group data and perform aggregations - sum, mean, count, min, max on the movies dataset.

```python
import pandas as pd 
df = pd.read_csv('movies_cleaned.csv')
df.shape
```

```python
df.groupby('original_language')['vote_average'].mean().sort_values(ascending=False).head(10)
```

Exercise 1 — Count of movies per language

```python
#df.groupby('original_language').size().sort_values(ascending=False).head(10) takes null value also
df.groupby('original_language')['title'].count().sort_values(ascending=False).head(10)
```

Exercise 2 — Total revenue per language

```python
pd.set_option('display.float_format', lambda x: f'{x:,.0f}')
df.groupby('original_language')['revenue'].sum().sort_values(ascending=False).head(10)
```

```python
pd.reset_option('display.float_format')
df.groupby('original_language')['vote_average'].agg(['mean', 'min', 'max', 'count']).sort_values('count', ascending=False).head(10)
```

```python
df.groupby(['original_language', 'status'])['revenue'].sum().sort_values(ascending=False).head(10)
```

```python
pd.set_option('display.max_rows', 20)
df.groupby('original_language').agg(
    total_revenue = ('revenue', 'sum'),
    avg_voting=('vote_average', 'mean'),
    movie_count = ('title', 'count')
).sort_values('movie_count', ascending=False).head(10)
```

```python
df.groupby('original_language').agg(
    total_revenue=('revenue', 'sum'),
    avg_voting=('vote_average', 'mean'),
    movie_count=('title', 'count')
).sort_values('movie_count', ascending=False).head(10).reset_index()
```
