## Filtering, Querying and Cleaning Data
Boolean filtering, query(), handling duplicates, handling nulls, and string filtering on the movies dataset.

```python
import pandas as pd
import numpy as nu

df = pd.read_csv('movies_cleaned.csv')
df.shape
```

### 1. Boolean Filtering
Filter rows based on conditions — similar to SQL WHERE clause

```python
# Movies with vote_average > 8
high_rated = df[df['vote_average'] > 8]
high_rated[['title', 'vote_average', 'genres']].head(10)
```

```python
# Movies with vote_average > 8 AND revenue > 100 million
blockbusters = df[(df['vote_average'] > 8) & (df['revenue'] > 100000000)]
blockbusters[['title', 'vote_average', 'revenue']].head(10)
```

```python
# Movies in English OR Hindi
en_hi = df[(df['original_language'] == 'en') | (df['original_language'] == 'hi')]
en_hi['original_language'].value_counts()
```

```python
# Movies in English, Hindi or Tamil — cleaner than multiple OR conditions
selected = df[df['original_language'].isin(['en', 'hi', 'ta'])]
selected['original_language'].value_counts()
```

```python
# Movies NOT in English
non_english = df[~(df['original_language'] == 'en')]
non_english['original_language'].value_counts().head(10)
```

### 2. Query Method
A cleaner, SQL-like way to filter — easier to read than boolean filtering

```python
df.query('vote_average > 8')[['title', 'vote_average', 'genres']].head(10)
```

```python
# Multiple conditions
df.query('vote_average > 8 and revenue > 100000000')[['title', 'vote_average', 'revenue']].head(10)
```

```python
# Using variables in query with @
min_rating = 7.5
min_revenue = 50000000
df.query('vote_average > @min_rating and revenue > @min_revenue')[['title', 'vote_average', 'revenue']].head(10)
```

### 3. String Filtering
Filter based on string content using .str accessor

```python
# Movies with 'love' in title (case insensitive)
df[df['title'].str.contains('love', case=False, na=False)][['title', 'genres']].head(10)
```

```python
# Movies whose title starts with 'The'
df[df['title'].str.startswith('The', na=False)][['title']].head(10)
```

```python
# Movies whose title is all uppercase
df[df['title'].str.isupper()][['title']].head(10)
```

### 4. Handling Nulls
Identifying, filling and dropping null values

```python
# Count nulls in each column
df.isnull().sum()
```

```python
# Percentage of nulls per column
(df.isnull().sum() / len(df) * 100).round(2)
```

```python
# Drop rows where title is null
df_clean = df.dropna(subset=['title'])
print(f"Before: {len(df)} rows")
print(f"After: {len(df_clean)} rows")
```

```python
# Fill null values
df_clean['overview'] = df_clean['overview'].fillna('No overview available')
df_clean['runtime'] = df_clean['runtime'].fillna(df_clean['runtime'].median())
df_clean[['title', 'overview', 'runtime']].head(10)
```

### 5. Handling Duplicates

```python
# Check for duplicates
print(f"Total rows: {len(df)}")
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Check duplicates based on specific column
print(f"Duplicate titles: {df.duplicated(subset=['title']).sum()}")
```

```python
# Drop duplicates keeping first occurrence
df_dedup = df.drop_duplicates(subset=['title'], keep='first')
print(f"After dedup: {len(df_dedup)} rows")
```

### 6. Sorting

```python
# Sort by single column
df.sort_values('vote_average', ascending=False)[['title', 'vote_average']].head(10)
```

```python
# Sort by multiple columns
df.sort_values(['vote_average', 'revenue'], ascending=[False, False])[['title', 'vote_average', 'revenue']].head(10)
```

```python
df_clean.to_csv('movies_filtered_clean.csv', index=False)
print("Saved!")
```

### Summary
- Boolean filtering with &, |, ~
- .query() for cleaner SQL-like filtering
- .str accessor for string operations
- Handling nulls with isnull(), fillna(), dropna()
- Handling duplicates with duplicated(), drop_duplicates()
- Sorting with sort_values()
