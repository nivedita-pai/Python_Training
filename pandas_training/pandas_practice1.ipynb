## Section 1 - Creating DataFrames from Dictionary and DateTime Operations
Creating a DataFrame from a Python dictionary and performing DateTime operations - extracting year, month, day, leap year, week of year and calculating age.

```python
import pandas as pd

pd.__version__
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'DOB' : ['1990-01-01', '1985-05-15', '1992-07-20', '1988-03-10', '1995-12-25'],
    'Score': [85, 90, 78, 92, 88]
})
df.info()
```

```python
df.describe()
```

```python
df.head(5)
```

```python
#df['year_of_birth']=df['DOB'].str[:4]

df['year_of_birth']=pd.to_datetime(df['DOB']).dt.year
df['month_of_birth']=pd.to_datetime(df['DOB']).dt.month
df['day_of_birth']=pd.to_datetime(df['DOB']).dt.day

df['is_leap_year']=pd.to_datetime(df['DOB']).dt.is_leap_year
df['day_of_week']=pd.to_datetime(df['DOB']).dt.day_name()

df.head(5)
```

Finding Age using numpy

```python
import numpy as np
today = pd.to_datetime('today')
year = today.year
date = today.day
month = today.month

birthday_date = pd.to_datetime(df['DOB']).dt.day
birthday_month = pd.to_datetime(df['DOB']).dt.month
birthday_year = pd.to_datetime(df['DOB']).dt.year

df['age'] = np.where(
    (month > birthday_month) | ((month == birthday_month) & (date >= birthday_date)),
    year - birthday_year,
    year - birthday_year - 1
)
df.head(5)
```

Finding Age using apply with regular Python function

```python
from datetime import date

def calculate_age(dob):
    today = date.today()
    dob = dob.date()
    age = today.year - dob.year
    if (today.month, today.day) < (dob.month, dob.day):
        age -= 1
    return age

df['age'] = pd.to_datetime(df['DOB']).apply(calculate_age)
df.head()
```

Exercise 2 — Apply on multiple columns using axis=1:
Add a column that combines Name and age into a sentence:

```python
df['student_details'] = df.apply(lambda x : f"{x['Name']} is {x['age']} years old", axis=1)

df.loc[:, 'student_details']
```

Exercise 3 — Apply with a named function:
Write a function that categorises age:

Below 35 → "Young"
35 to 40 → "Mid"
Above 40 → "Senior"

```python
def student_category(age):
    if age <35:
        return 'Young'
    elif age <40 :
        return 'Middle-aged'
    else:
        return 'Senior'
    
df['age_category']=df['age'].apply(student_category)
df.loc[:, ['Name', 'age', 'age_category']]
```
