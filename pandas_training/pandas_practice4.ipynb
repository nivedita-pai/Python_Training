### Pandas Practice 4
Merging And Joining

Learn to merge and join 

```python
import pandas as pd

# DataFrame 1 - Customer orders
orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4, 5],
    'customer_id': [101, 102, 103, 101, 104],
    'amount': [250, 150, 300, 200, 175]
})

# DataFrame 2 - Customer details
customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'city': ['Chennai', 'Mumbai', 'Delhi', 'Bangalore']
})

print(orders)
print(customers)
```

```python
inner = pd.merge(orders, customers, on='customer_id', how='inner')
inner
```

```python
left = pd.merge(orders, customers, on='customer_id', how='left')
left
```

```python
right= pd.merge(orders, customers, on='customer_id', how='right')
right
```

```python
outer = pd.merge(orders,customers,on='customer_id',how='outer')
outer
```
